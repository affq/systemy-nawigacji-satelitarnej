```python
import matplotlib.pyplot as plt
import numpy as np
from fun import Hirvonen, Rneu

xyzref = [3655333.847, 1403901.067, 5018038.047]
blhref = Hirvonen(xyzref[0], xyzref[1], xyzref[2])
rneu = Rneu(blhref[0], blhref[1])

interval = 30
```

## CBKA

```python
CBKA = np.genfromtxt(r'C:\Users\adria\Desktop\systemy-nawigacji-satelitarnej-1\projekt3\wyniki\CBKA.pos', comments='%') # wektor krtki
dxyz_CBKA = CBKA[:,2:5] - xyzref
dneu_CBKA = []

for i, xyz in enumerate(dxyz_CBKA):
    neu = rneu.T.dot(xyz)
    dneu_CBKA.append([i*interval/3600, neu[0], neu[1], neu[2]])

plt.figure()
plt.plot([d[0] for d in dneu_CBKA], [d[1] for d in dneu_CBKA], label='dN')
plt.plot([d[0] for d in dneu_CBKA], [d[2] for d in dneu_CBKA], label='dE')
plt.plot([d[0] for d in dneu_CBKA], [d[3] for d in dneu_CBKA], label='dU')
plt.xlabel('czas [h]')
plt.ylabel('bd [m]')
plt.title('Bd wsprzdnych dla CBKA')
plt.legend()
plt.grid()
plt.show()

fig, axes = plt.subplots(3, 1)
axes[0].plot([d[0] for d in dneu_CBKA], [d[1] for d in dneu_CBKA])
std = np.std([d[1] for d in dneu_CBKA])
rms = np.sqrt(np.mean([d[1]**2 for d in dneu_CBKA]))
max = np.max([np.abs(d[1]) for d in dneu_CBKA])
axes[0].text(1.5, 0.95, f" = {std:.3f}m \n RMS = {rms:.3f}m \n d_max = {max:.3f}m", transform=axes[0].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
axes[0].set_title('dN')
axes[0].set_xlabel('czas [h]')
axes[0].set_ylabel('bd [m]')
axes[0].grid()

axes[1].plot([d[0] for d in dneu_CBKA], [d[2] for d in dneu_CBKA], '#ff7f0e')
std = np.std([d[2] for d in dneu_CBKA])
rms = np.sqrt(np.mean([d[2]**2 for d in dneu_CBKA]))
max = np.max([np.abs(d[2]) for d in dneu_CBKA])
axes[1].text(1.5, 0.95, f" = {std:.3f}m \n RMS = {rms:.3f}m \n d_max = {max:.3f}m", transform=axes[1].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
axes[1].set_title('dE')
axes[1].set_xlabel('czas [h]')
axes[1].set_ylabel('bd [m]')
axes[1].grid()

axes[2].plot([d[0] for d in dneu_CBKA], [d[3] for d in dneu_CBKA], '#2ca02c')
std = np.std([d[3] for d in dneu_CBKA])
rms = np.sqrt(np.mean([d[3]**2 for d in dneu_CBKA]))
max = np.max([np.abs(d[3]) for d in dneu_CBKA])
axes[2].text(1.5, 0.95, f" = {std:.3f}m \n RMS = {rms:.3f}m \n d_max = {max:.3f}m", transform=axes[2].transAxes, ha='right', va='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
axes[2].set_title('dU')
axes[2].set_xlabel('czas [h]')
axes[2].set_ylabel('bd [m]')
axes[2].grid()

plt.tight_layout()
plt.show()

fix = CBKA[:,5] == 1
floats = CBKA[:,5] == 2

fig, ax = plt.subplots()
ax.bar('fix', np.sum(fix), color='#962a69')
ax.bar('float', np.sum(floats), color='#5fb6c7')
ax.text(0, np.sum(fix), str(np.sum(fix)), ha='center', va='bottom')
ax.text(1, np.sum(floats), str(np.sum(floats)), ha='center', va='bottom')
ax.set_title('Liczba poszczeglnych typw rozwiza wektora')

plt.show()
```
